```python
import pandas as pd
from numpy import NaN
from tqdm import tqdm
from pororo import Pororo
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data_path = '/content/gdrive//MyDrive/Stage2/input/data/train/'
mt = Pororo(task="translation", lang="multi")
dataset = pd.read_csv(data_path+'train.tsv', delimiter='\t', header=None)
```

```python
def cycling_translation_zh(row) :
    s = str(row[1])
    s_zh = mt(s, 'ko', 'zh')
    s_ko = mt(s_zh, 'zh', 'ko')
    if str(row[2]) in s_ko and str(row[5]) in s_ko :
        return s_ko
    else :
        return NaN
```

```python
new_sentences = []
for idx, row in tqdm(dataset.iterrows(), total=len(dataset)) :
    new_sentences.append(cycling_translation_zh(row))
```

```python
dataset[1] = new_sentences
```

```python
dataset.to_csv(data_path+'new_data_NaN_zh.tsv', sep='\t', index=False, header=False)
```

```python
dataset = dataset.dropna()
dataset.to_csv(data_path+'new_data_zh.tsv', sep='\t', index=False, header=False)
```
